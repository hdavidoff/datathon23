```python
import pandas as pd
import translators as ts
from googletrans import Translator
translator = Translator()
import tqdm
from collections import Counter
```

```python
input_file = "invaluable_prices.csv"
output_file = "invaluable_english.csv"
#prices_df = pd.read_csv(input_file)
#prices_df["name_english"] = None
#prices_df.to_csv(output_file, index=None)
prices_df = pd.read_csv(output_file)
prices_df
```

```python
counter = 0
indices = prices_df[prices_df["name_english"].isna()].index.values.astype(int)
for index in tqdm.tqdm(indices):
    counter += 1
    if counter%100 == 0:
        prices_df.to_csv(output_file, index=None)
    try:
        prices_df.at[index, 'name_english'] = translator.translate(prices_df.at[index, 'name']).text
    except:
        prices_df.at[index, 'name_english'] = ts.translate_text(prices_df.at[index, 'name'], to_language='en')
prices_df.to_csv(output_file, index=None)
```
